```python
Omax = tesla['Open'].max()
Omin = tesla['Open'].min()

tesla['OpenNorm'] = tesla['Open'].apply(lambda x: (x - Omin)/(Omax - Omin))

x = tesla['Date']
y = tesla['OpenNorm']


plt.xlabel('Fecha')

plt.title('Date Vs. OpenNorm.')

plt.xticks([0, 1288, 2576])

plt.plot(x, y, color = 'r', label = 'Open')

plt.show()
```
